```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import os 
from skimage import io 
from matplotlib.patches import Rectangle
```

```python
# Extracting images

root_path = 'C:/Users/jonas.peterschmitt/RiderProjects/data/SignverOD/images/'
filenames = os.listdir(root_path)

# Extracting imageId for image chosen

rootImagesId = 'C:/Users/jonas.peterschmitt/RiderProjects/data/SignverOD/'
Images_df = pd.read_csv(rootImagesId+'image_ids.csv')

#Extracting boxes in the document

Boxes_df = pd.read_csv(rootImagesId+'train.csv')
Boxes_df.head()
```

```python
#fig, axs = plt.subplots(15, 3)


for i in range(14,30):
    plt.figure(figsize=(15,10))
    image_test = io.imread(root_path+filenames[i])
    ImageId = Images_df[Images_df['file_name']==filenames[i]]
    
    plt.imshow(image_test)
    
    Boxes = Boxes_df[Boxes_df['image_id']==int(ImageId['id'])].index
    Index_div = Boxes[:-1]+1==Boxes[1:]
    if False in Index_div : 
        separation = Index_div.tolist().index(False)
        Boxes_1 = Boxes_df.iloc[Boxes[:separation+1]]['bbox']
    else :
        Boxes_1 = Boxes_df[Boxes_df['image_id']==int(ImageId['id'])]['bbox']
        
    for coord in Boxes_1 : 
        coord = coord[1:len(coord)-1].split(', ')
        coord = [float(i) for i in coord]
        currentAxis = plt.gca()
        xa,xb,ya,yb = coord
        currentAxis.add_patch(Rectangle((xa*float(ImageId['width']),xb*float(ImageId['height'])),  ya*float(ImageId['width']), yb*float(ImageId['height']) , fill=None, alpha=1))
    plt.show()
    
```
